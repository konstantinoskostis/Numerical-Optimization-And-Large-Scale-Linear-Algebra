## Minimizing Risk (Squared) with target (annualized) return value of $ρ$

According to the Readme PDF, the following equation (17.2):

$minimize ||Rw - ρ\boldsymbol{1}||^2$ (1)

s.t 

$\sum_{i=1}^{n} w = 1$ (2) and \
$\mu^Tw = ρ$ (3)

defines a linearly constrained least squares problem, for the portfolio optimization.

We care about the value `w` which comes from the solution of these equations.

- In our case:
  - $R$ is the R_train .
  - $w$ is the weight vector of the 20 assets.
  - $ρ$ will take the value of `10%`, `20%` and `40%` as requested by the exercise
  - $\mu = R^T \boldsymbol{1}/T$ is the n-vector (where n=20) of the average asset returns
      - $T$ is the number of periods (in our case the number of rows/days in the dataset)


**Methodology**

A class named `PortfolioOptimizer` is created in order to find the weights/allocations for a given target annualized return \
and encodes all operations (risk-minimization, cummulative portfolio value over time calculation, leverage) needed as building blocks \
for the tasks of the assignment.

The class `PortfolioOptimizer` contains:

- `fit(p, P=250)`: minimizes the (squared) risk, with `scipy.optimize.minimize`. The output of this method
  is the allocation (weights for assets) for a given `p` (annualized target return). It simply implements (1)
  using (2) and (3), as a linearly constraint least squares problem.
  - The method also allows for `allow_short_selling` and $\lambda$ parameters.
- `predict(R)`: implements $r = Rw$ where $r$ is the `T-vector` of portfolio returns in the t periods.
- `V(R, investment_init)`: implements/calulates the cummulative value of a portfolio over time using the formula 17.1, which is:
  - $V_t = V_1(1+r_1)(1+r_2)...(1+r_{t-1})$
- `L`: computes the leverage using the formula $L = \sum_{i=1}^{n} |w_i|$

```python
# The number of trading days is 250
P = 250
```

```python
class PortfolioOptimizer:
    """Portfolio optimizer"""

    def __init__(self, A):
        self.A = A
        self.w = np.array([])

    def fit(self, p, P=P, allow_short_selling=True, lamda = 0):
        """ Find the allocation (weights) via portfolio risk minimization """

        # The objective function (as per the Readme file)
        # Essentially minimizes the risk (portfolio variance)
        def objective_function(weights):
            res = np.dot(self.A, weights) - p
            risk = np.dot(res, res)

            if lamda == 0:
                return risk
            else:
                # keep only variances and weights for the risky assets
                variances = self.A.std(axis=0)[:-1]**2
                reg_weights = weights[:-1]**2

                regularized = risk + lamda*np.dot(variances, reg_weights)
                return regularized

        # The sum to one of weights, constraint
        sum_to_one_constraint = {
            'type': 'eq', 'fun': lambda w: np.sum(w) - 1
        }

        # The mean portofolio return (annualized) of equal to p (target), constraint
        daily_returns = self.A
        mean_portfolio_return_constraint = {
            'type': 'eq', 'fun': lambda w: w.dot(daily_returns.mean(axis=0)) * P - p
        }

        # Gather the constraints to a tuple
        constraints = (sum_to_one_constraint, mean_portfolio_return_constraint)

        # Enable the no-short sales constraint if needed
        # Meaning that each weight is >= 0 (with no upper bound)
        if not allow_short_selling:
            bounds = [(0, None) for i in range(self.A.shape[1])]
        else:
            bounds = None

        # Initial guess for the weights (random with normal distribution, mean=0, std=1)
        w_init = np.random.normal(loc=0, scale=1, size=self.A.shape[1])

        # Minimize the objective function subject to the above linear constraints
        result = scipy.optimize.minimize(
            fun = objective_function,
            x0 = w_init,
            constraints=constraints,
            bounds=bounds,
            method='SLSQP',
            tol=1e-6
        )

        # Obtain x (the solution/weights)
        self.w = result.x

        return self

    def predict(self, R):
        """
          Predict the return time series: r

          r = Rw (matrix to vector multiplcation)
        """
        return np.dot(R, self.w)

    def V(self, R, investment_init):
        """ Cummulative portfolio value calculation over time """
        r = self.predict(R)
        cummulative_weighted_returns = (1 + r).cumprod()
        v = investment_init * cummulative_weighted_returns

        return v

    def L(self):
        return np.sum(np.abs(self.w))
```

## 1.a) Plot cummulative values for training set for 10%, 20% and 40% portfolio

**Methodology**

- Calculate the cummulative portfolio values over time, using the method `V(R)` of the `PortfolioOptimizer` class (The y-axis)
- Put the time (days) in the x-axis
- Plot the values

This is realized by creating a helper/reusable function for plotting.

```python
def plot_portfolio_value(R, investment_init, optimizers, colors, dataset_label):
    # Calculate the cummulative portfolio values over time
    portfolio_values = {}

    for optimizer_key in optimizers.keys():
        portfolio_values[optimizer_key] = optimizers[optimizer_key].V(R, investment_init)

    # Arange the days
    days = np.arange(0, R.shape[0])
    
    # Plot the values
    plt.figure(figsize=(8, 6))

    title_returns = ', '.join(list(optimizers.keys()))
    title = "Cumulative values for {} returns on {} set".format(title_returns, dataset_label)
    
    for optimizer_key in portfolio_values:
        plt.plot(days, portfolio_values[optimizer_key], color=colors[optimizer_key], label=optimizer_key)

    plt.xlabel('Day')
    plt.ylabel('Value')

    plt.title(title)
    
    plt.legend()
    plt.show()
```

```python
# Plot for the train set
optimizers_map = {
    '10%': optimizer_10,
    '20%': optimizer_20,
    '40%': optimizer_40
}

colors = {
    '10%': 'brown',
    '20%': 'green',
    '40%': 'black'
}

plot_portfolio_value(R_train, 10_000, optimizers_map, colors, 'training')
```

## 1.b) Plot cummulative values for test set for 10%, 20% and 40% portfolio

```python
# Plot for the test set

plot_portfolio_value(R_test, 10_000, optimizers_map, colors, 'test')
```

## 2. Portfolio Report

In this section, we report for each portfolio (10%, 20% and 40%) for both train and test datasets:
- the annual return
- the annual risk
- the asset with the minimum allocation
- the asset with the maximum allocation
- the leverage

Reporting is realized as pandas Dataframes, using a custom class named `PortfolioReport` to achieve reusability.

```python
class PortfolioReporter:
    def __init__(self, dataset, dataset_label, optimizers, stock_names, P=P):
        self.dataset = dataset
        self.dataset_label = dataset_label
        self.optimizers = optimizers
        self.stock_names = stock_names
        self.P = P

    def report(self):
        report_df = pd.DataFrame()

        report_df['Dataset'] = [self.dataset_label] * len(self.optimizers)
        report_df['Portfolio'] = list(self.optimizers.keys())

        report_df['Annual Return'] = [
            np.mean(self.optimizers[optimizer_key].predict(self.dataset)) * self.P
            for optimizer_key in self.optimizers
        ]

        report_df['Annual Risk'] = [
            np.std(self.optimizers[optimizer_key].predict(self.dataset)) * np.sqrt(self.P)
            for optimizer_key in self.optimizers
        ]

        report_df['Asset with Minimum Allocation'] = [
            self.stock_names[np.argmin(self.optimizers[optimizer_key].w)]
            for optimizer_key in self.optimizers
        ]

        report_df['Minimum Allocation'] = [
            np.min(self.optimizers[optimizer_key].w)
            for optimizer_key in self.optimizers
        ]

        report_df['Asset with Maximum Allocation'] = [
            self.stock_names[np.argmax(self.optimizers[optimizer_key].w)]
            for optimizer_key in self.optimizers
        ]

        report_df['Maximum Allocation'] = [
            np.max(self.optimizers[optimizer_key].w)
            for optimizer_key in self.optimizers
        ]

        report_df['Leverage'] = [
            self.optimizers[optimizer_key].L()
            for optimizer_key in self.optimizers
        ]

        return report_df
```

```python
# Construct a dataframe for portfolios based on the train set
# using the PorfolioReporter class

train_reporter = PortfolioReporter(R_train, 'Train', optimizers_map, stock_names_map)
train_reporter.report()
```

```python
# Construct a dataframe for portfolios based on the test set
# using the PorfolioReporter class

test_reporter = PortfolioReporter(R_test, 'Test', optimizers_map, stock_names_map)
test_reporter.report()
```

## Extra: Experiment with regularization for the 40% return portfolio

In this section regularization is added to the objective according to the formula proposed in the variations sections of the Readme file.

The regularization term will take the form: $\lambda*(\sum_{i=1}^{n-1} \sigma_i*w_i^2)$ (and will be added to the objective function)

The weight of the risk-free-asset is not being regularized!

**Methodology**
- Split R_train to 80% (R_t) and 20% (R_v) 
- Fine tune $\lambda$ for a list of 400 values evenly split from $10^{-2}$ to $10^2$
- On each $\lambda$ train on R_t set and predict on R_v set. Measure the absolute error from the target return (0.4)
- For the best $\lambda$ train on the full train set R_train.
- Report (graphs and pandas) on both R_train and R_test.

```python
class LamdaTuner:
    def __init__(self, lamda_list):
        self.lamda_list = lamda_list
        self.errors = []

    def tune(self, target_return, R_t, R_v, P=P):
        for lamda in self.lamda_list:
            # train using set R_t
            optimizer = PortfolioOptimizer(R_t).fit(target_return, P=P, lamda=lamda)
            # predict on set R_v
            r = optimizer.predict(R_v)
            y_predicted = np.dot(r, r)
            error = np.abs(target_return - y_predicted)
            self.errors.append(error)
```

```python
# Lamda list in logarithmic scale: from 10^-2 to 10^2, 400 values of lamda 
lamdas = np.logspace(-2, 2, 400)

# Split the train set into 80%/20%
R_t, R_v = train_test_split(R_train, test_size=0.2, random_state=seed)

tuner = LamdaTuner(lamdas)
tuner.tune(0.4, R_t, R_v)
```

```python
best_lamda = lamdas[np.array(tuner.errors).argmin()]
print("Best lamda is: {}".format(best_lamda))
```

```python
# Re-fit using lamda
reg_optimizer = PortfolioOptimizer(R_train).fit(0.4, lamda=best_lamda)
optimizers_reg = {
    '40%': reg_optimizer
}
```

```python
train_reporter = PortfolioReporter(R_train, 'Train (with regularization)', optimizers_reg, stock_names_map)
train_reporter.report()
```
